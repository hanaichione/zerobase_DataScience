### 연습문제
- makedata 모듈을 이용해서 데이터 프레임 만들기
- user_df
    - 8명의 데이터가 있는 데이터 프레임을 만드시오
    - UserID
    - Name : makedata.get_name()
    - Age : makedata.get_age()
    - 중복되는 Name 값이 없도록

```python
Name = []
Age = []

while True:
    name_data = makedata.get_name()
    age_data = makedata.get_age()
    if name_data not in Name:
        Name.append(name_data)
        Age.append(age_data)
    elif len(Name) == 8:
        break;
    else:
        continue
```

```python
# 선생님 풀이

# 1. 딕셔너리의 리스트

datas = {}
datas["UserID"] = list(range(1, 9))
datas["Age"] = [makedata.get_age() for _ in range(8)]
# 8만큼 for 문을 돌려서 데이터를 받긴 하는데 사용하지 않을 때 '_'

names = []
while True:
    name = makedata.get_name()
    if name not in names:
        names.append(name)
    if len(names) >= 8:
        break

datas["Name"] = names
datas
user_df = pd.DataFrame(datas)
user_df
```

```python
# 리스트의 딕셔너리

user_df = pd.DataFrame(columns=["UserID", "Name", "Age"]) # 컬럼만 있는 데이터 프레임 생성

for idx in range(1, 9):
    name = makedata.get_name()
    while name in list(user_df["Name"]):
        name = makedata.get_name()
    
    data = {"Name":name, "UserID":idx, "Age":makedata.get_age()}
    
    user_df.loc[len(user_df)] = data
    
user_df
```

### 연습문제
- money_df 만들기
    - 15개의 데이터
    - ID : 1 ~ 8 랜덤한 숫자 데이터
    - Money : 1000원 단위로 1000원 ~ 20000원까지의 숫자가 저장

```python
# 딕셔너리 데이터를 데이터 프레임에 하나씩 추가하기
money_df = pd.DataFrame(columns = ["ID", "Money"])

for _ in range(1, 16):
    money_df.loc[len(money_df)] = {
        "ID":np.random.randint(1, 9), 
        "Money":1000*(np.random.randint(1, 21))
    }
# 컬럼데이터에서 Unique 값 확인

ids = money_df["ID"].unique()
ids.sort()
ids
```

### 1. Merge

```python
user_df.merge(money_df, left_on="UserID", right_on="ID")
# left_on, right_on 키값
```

```python
# 컬럼명 변경
user_df.rename(columns={"UserID":"ID"}, inplace=True)
user_df.tail(1)
```

```python
user_df.merge(money_df).tail(2) # 데이터 프레임 객체 내에 있는 merge 사용
# left_on, right_on 쓸 필요 없음
```

```python
result_df = pd.merge(money_df, user_df) # 판다스 내에 있는 merge 사용
result_df
```

```python
# groupby : sum, size, min... 함수 : Series

money_list = result_df.groupby("Name").sum()["Money"].reset_index() # inner로 돼서 없는 건 안 나옴
money_list
```

```python
# groupby : agg("sum"), agg("mean")... : DataFrame

money_list2 = result_df.groupby("Name").agg("sum").reset_index()[["Name", "Money"]] 
# [["Name", "Money"]]로 마스킹 해 줘야 해당되는 두 가지만 나옴
money_list2
```

```python
# merge : money_list, user_df : outer
```

```python
result = pd.merge(user_df, money_list, how="outer")
```

```python
result
```
