```python
# 베이지안 추정 및 디리클레 분포 그리기

def project(x):
    n1 = np.array([1, 0, 0])
    n2 = np.array([0, 1, 0])
    n3 = np.array([0, 0, 1])
    n12 = (n1 + n2) / 2
    m1 = np.array([1, -1, 0])
    m2 = n3 - n12
    m1 = m1 / np.linalg.norm(m1)
    m2 = m2 / np.linalg.norm(m2)
    return np.dstack([(x - n12).dot(m1), (x - n12).dot(m2)])[0]

def project_reverse(x):
    n1 = np.array([1, 0, 0])
    n2 = np.array([0, 1, 0])
    n3 = np.array([0, 0, 1])
    n12 = (n1 + n2) / 2
    m1 = np.array([1, -1, 0])
    m2 = n3 - n12
    m1 = m1 / np.linalg.norm(m1)
    m2 = m2 / np.linalg.norm(m2)
    return x[:, 0][:, np.newaxis] * m1 + x[:, 1][:, np.newaxis] * m2 + n12

eps = np.finfo(float).eps * 10
X = project([[1 - eps, 0, 0], [0, 1 - eps, 0], [0, 0, 1 - eps]])

import matplotlib.tri as mtri
triang = mtri.Triangulation(X[:, 0], X[:, 1], [[0, 1, 2]])
refiner = mtri.UniformTriRefiner(triang)
triang2 = refiner.refine_triangulation(subdiv=6)
XYZ = project_reverse(np.dstack([triang2.x, triang2.y, 1 - triang2.x - triang2.y])[0])

pdf = sp.stats.dirichlet((3, 5, 2)).pdf(XYZ.T)
plt.tricontourf(triang2, pdf, cmap=plt.cm.bone_r)
plt.axis("equal")
plt.show()
```

```python
# 베타 분포의 모수 추정
np.random.seed(0)
x = sp.stats.beta(15, 12).rvs(10000)

sns.distplot(x, kde=False, norm_hist=True)
plt.title("beta distribution sample histogram")
plt.show()
```

```python
def estimate_beta(x):
    x_bar = x.mean()
    s2 = x.var()
    a = x_bar * (x_bar * (1 - x_bar) / s2 - 1)
    b = (1 - x_bar) * (x_bar * (1 - x_bar) / s2 - 1)
    return a, b

params = estimate_beta(x)
print(params)
```

```python
# 확률밀도함수 추정하여 나타내기
xx = np.linspace(0, 1, 1000)
sns.distplot(x, kde=False, norm_hist=True)
plt.plot(xx, sp.stats.beta(params[0], params[1]).pdf(xx))
plt.xlim(0, 1)
plt.title("beta histogram & pdf")
plt.show()
```

```python
# distplot의 fit 인수로 모수를 추정하고 확률밀도함수 그리기
sns.distplot(x, kde=False, norm_hist=True, fit=sp.stats.beta)
plt.xlim(0, 1)
plt.title("beta histogram & pdf")
plt.show()
```

```python
# 상품 A와 B의 베르누이 모수의 사후분포 확률밀도함수 그리기
xx = np.linspace(0, 1, 1000)
a0, b0 = 3, 2
plt.plot(xx, sp.stats.beta(a0, b0).pdf(xx), c='r', ls='--', label="A pdf")
a1, b1 = 61, 41
plt.plot(xx, sp.stats.beta(a1, b1).pdf(xx), c='b', ls='-', label="B pdf")
plt.legend()
plt.title("pdf by Beyesian estimation")
plt.show()
```

### 3. 독립표본 t 검정

```python
N_1 = 10
mu_1 = 0
sigma_1 = 1
N_2 = 10
mu_2 = 0.5
sigma_2 = 1

np.random.seed(0)
x1 = sp.stats.norm(mu_1, sigma_1).rvs(N_1)
x2 = sp.stats.norm(mu_2, sigma_2).rvs(N_2)

ax = sns.distplot(x1, kde=False, fit=sp.stats.norm, label="No.1")
ax = sns.distplot(x2, kde=False, fit=sp.stats.norm, label="No.2")
ax.lines[0].set_linestyle(":")
plt.legend()
plt.show()
```

```python
np.mean(x1), np.mean(x2)
```

```python
sp.stats.ttest_ind(x1, x2, equal_var=False)
```

```python
x1 = [80, 75, 85, 50, 60, 75, 45, 70, 90, 95, 85, 80]
x2 = [80, 85, 70, 35, 55, 80]
sp.stats.ttest_ind(x1, x2, equal_var=True)
```

### 4. 대응표본 t검정

```python
N = 5
mu1 = 0
mu2 = 0.4
np.random.seed(1)
x1 = sp.stats.norm(mu1).rvs(N)
x2 = x1 + sp.stats.norm(mu2, 0.1).rvs(N)

ax = sns.displot(x1, kde = False, fit=sp.stats.norm, label="n.1")
ax = sns.displot(x2, kde = False, fit=sp.stats.norm, label="n.2")
ax.lines[0].set_linestyle(":")
plt.legend()
plt.show()
```
