```python
total = np.hstack((ll.data, ll.target))
```

```python
total
```

```python
pca = PCA(n_components=1)
W = pca.fit_transform(total)
W
```

```python
recovered_data = pca.inverse_transform(W)
recovered_data
```

```python
print("%0.2f"%(pca.explained_variance_ratio_))
```
