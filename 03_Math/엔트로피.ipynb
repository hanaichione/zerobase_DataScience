# 엔트로피

```python
# 각각의 확률값에 대해 엔트로피를 계산하여 그래프 그리기

# log에 0넣으면 오류나므로 가장 작은 수
p0 = np.linspace(0.001, 1-0.001, 1000)
p1 = 1-p0
h = -p0*np.log(p0) - p1*np.log2(p1)

plt.plot(p1, h, '-', label="entropy")
plt.legend()
plt.xlabel('p(y=1)')
plt.show()
```

```python
# 엔트로피 계산하기

p1 = [1/3, 2/3]
h1 = sp.stats.entropy(p1, base=2)

p2 = [3/4, 1/4]
h2 = sp.stats.entropy(p2, base=2)

p3 = [0, 1]
h3 = sp.stats.entropy(p3, base=3)

print(h1, h2, h3)
```
